## Building a popularity based model

```python
## Converting date column to datetime format and selecting only the ratings since 2017 because ---
## --- mobile phones become outdated in 2 years

new_df['date'] = pd.to_datetime(new_df['date'])
pop_reco_df = new_df[new_df['date']>'2017-01-01']
```

```python
## picking up mobile phones with atleast 200 ratings since popularity will not only depend on average ratings but also on ---
## --- the number of ratings given

lst3 = pop_reco_df['product'].value_counts()[pop_reco_df['product'].value_counts()>=200].index.values
pop_reco_df = pop_reco_df.query('product in @lst3')
```

```python
## Final recommendation of five most popular phones

pop_reco_df.groupby('product')['score'].mean().sort_values(ascending = False).head()
```

## Building a collaborative model with SVD and KNNwithMeans 

```python
## Laoding the required libraries

from surprise import SVD
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
## Reducing size of database based on latest ratings only (from 1st Jan 2017). 
## Here we can also choose the dataset with products with atleast 50 ratings and users with atleast 50 ratings.
## I chose the larger dataset and filtered from there (since after multiple iterations final model accuracy was better for entire dataset) )

df['date'] = pd.to_datetime(df['date'])
usable_df = df[df['date']>'2017-01-01']
```

```python
usable_df.shape
```

```python
## Loading dataset from df to surprise library usable form

reader = Reader(rating_scale=(0, 10))
data = Dataset.load_from_df(usable_df[['author','product','score']], reader)
trainset, testset = train_test_split(data, test_size=.15, random_state=42)
```

```python
## Fitting the SVD model

model1 = SVD()
model1.fit(trainset)
```

```python
## Predictions with SVD model

model1_test_pred = model1.test(testset)
```

```python
## Displaying predictions

model1_test_pred
```

```python
## getting RMSE for SVD Model


print("SVD Model : Test Set")
accuracy.rmse(model1_test_pred, verbose=True)
```

```python
## Fitting the KNNwithMeans model (user based collab)

model2 = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
model2.fit(trainset)
```

```python
## Predictions with KNNwithMeans model (user based collab)

model2_test_pred = model2.test(testset)
```

```python
## Displaying predictions

model2_test_pred
```

```python
## getting RMSE for KNNwithMeans model (user based collab)


print("User based Model : Test Set")
accuracy.rmse(model2_test_pred, verbose=True)
```

```python
## Fitting the KNNwithMeans model (item based collab)

model3 = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': False})
model3.fit(trainset)
```

```python
## Predictions with KNNwithMeans model (item based collab)

model3_test_pred = model3.test(testset)
```

```python
## Displaying predictions

model3_test_pred
```

```python
## getting RMSE for KNNwithMeans model (item based collab)


print("Item based Model : Test Set")
accuracy.rmse(model3_test_pred, verbose=True)
```

```python
## Transferring test predictions from SVD Model to a dataframe

testusers_df = pd.DataFrame(model1_test_pred)
```

```python
## Average rating for test users

testusers_df.est.mean()
```

## Findings and Inferences

1. Amongst the Collaborative models SVD model performs the best in terms of accuracy with minimum RMSE of 2.39 vs 2.49 of item based model and 2.56 of user based model
2. In terms of time taken to run: item based model is much faster than the other two models followed by SVD and user based model (slowest)
3. As such the difference in rating estimations between the two KNNwithMeans models is not very high but item based model performs better (more accurate) as well as works faster.

## Recommending top 5 products for test users

```python
## creating a new testset to get predictions for test users on all products in the test set

users_array = testusers_df.uid.unique()
items_array = testusers_df.iid.unique()
newtestset = []
for i in users_array:
    for j in items_array:
        x = (i,j,0)
        newtestset.append(x)
```

```python
## Predicting the ratings for each product for each user

new_test_pred = model3.test(newtestset)
new_test_pred_df = pd.DataFrame(new_test_pred)
```

```python
## Getting the top five recommendations for 25 users. Limiting it to 25 to save runtime

new_test_pred_df = new_test_pred_df.sort_values('est',ascending = False)
recommendation = []
c=0
for i in users_array[0:25]:
    a = new_test_pred_df[new_test_pred_df['uid']==i].head().iid.values
    y = [i,a[0],a[1],a[2],a[3],a[4]]
    recommendation.append(y)
```

```python
## Displaying recommendations in a dataframe

reco_df = pd.DataFrame(recommendation,columns = ['UserName','Reco1','Reco2','Reco3','Reco4','Reco5'])
reco_df
```
